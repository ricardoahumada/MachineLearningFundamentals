# Modelos de arbol de decision

Los rboles de decisin son componentes bsicos de mtodos muy avanzados y potentes, como *Random Forests* y *Gradient Boosted trees*.
Esos mtodos son a menudo el mtodo de mejor rendimiento cuando se trabaja con conjuntos de datos tabulares.

## Qu es un rbold e decisin?

- Tomemos el dataset de censo y en concreto, los ingresos de la poblacin.



- Al mirar los datos, podramos obtener un conjunto de reglas de decisin para ayudarnos a clasificar a las personas en diferentes grupos de ingresos.



> Esta estructura jerrquica: reglas de decisin basadas en una variable es lo que llamamos un rbol de decisin en el aprendizaje automtico.

- Los rboles tienen **nodos**, como el nodo raz, que son nodos de decisin o nodos divididos porque dividen los datos en dos subconjuntos.

- Y los nodos finales se llaman **hojas** del rbol, los nodos de la hoja, y sostienen los valores de prediccin.

## Cmo hacer crecer un rbol de clasificacin?

1. Comenzamos con la raz, que divide el dataset mediante un umbral y se calcula la probabilidad de una clase:
    - El umbral maximizar la separacin de las clases.
    - Hay una cantidad que se llama **entropa o la puntuacin de Gini** que permite cuantificar qu tan buena es una divisin.
    - Y automticamente, el algoritmo encontrar el mejor valor umbral para maximizar la mejora en la entropa despus de la divisin.



2. Dividimos un nodo y lo convertimos en un nodo de decisin, aadiendo un umbral que maximiza la separacin.



3.Volvemos a repetir el proceso



Se Calcula la evolucin de la entropa antes y despus de la divisin y se encuentra la mejor divisin de esta manera.

A menudo es posible que un rbol de dos niveles como este no sea suficiente.

En conjuntos de datos ms complejos, podramos imaginar que tendramos que 'cultivar' un rbol ms profundo para poder separar perfectamente el conjunto de entrenamiento.

## Regresin con un rbol de decisin

Podemos aplicar las mismas ideas a los rboles de regresin.
- Esta vez, el objetivo ser predecir una variable continua en lugar de valores de clase.



1. Here again, we start with the root node, which is a decision node, a split.
- Nos vemos forzados a encontrar un valor de divisin para X, por lo que podemos dividir en toda la ubicacin posible de X.
- En el lado izquierdo de esa divisin, calcularemos **el valor promedio de la variable de destino Y para todos los puntos de datos que estn a la izquierda con respecto a la variable X**.
    - la lnea azul representa el valor promedio de todos los puntos en esa regin, a la izquierda de la divisin.
    - Esto se representa por el crculo azul oscuro para un valor fuerte de Y, y azul claro para un valor promedio ms bajo.
- La funcin de prediccin de un modelo de regresin con una sola divisin es esta funcin *step*, que es bsicamente constante por partes.



2. Podemos refinar este rbol de decisin, nuevamente, dividiendo los nodos para tomar decisiones adicionales.



3. Aadimos ms divisiones




> - Si introducimos ms y ms divisiones: cada vez que hacemos una nueva divisin, aadimos un nuevo nodo de hoja y cuantos ms nodos de hoja tengamos, ms niveles podremos predecir.
    - progresivamente, la funcin de prediccin se est comenzando a aproximar a la forma curvilnea del conjunto de entrenamiento.

So, how deep should we grow the trees?
Is it fine to go until the end?
Or should we stop before splitting each individual data point in the training set?
This is the topic of overfitting because as we will see, if we grow the tree too deep, then we would see that it impacts the generalization performance badly.

#### **Qu tan profundo deberamos hacer crecer los rboles**
- Est bien ir hasta el final? O deberamos detenernos antes de dividir cada punto de datos individual en el conjunto de entrenamiento?
- Ambito del **sobreajuste**: 
    - si hacemos crecer el rbol demasiado profundamente, veremos que afecta el rendimiento de la generalizacin a mal.

**Compensacin subajuste / sobreajuste**



> En Scikit-Learn, podemos decidir si deseamos controlar o no la profundidad mxima o el nmero mximo de nodos de hoja.
- Se tienen las dos posibilidades, y tienen un impacto equivalente en controlar la compensacin de subajuste-sobreajuste.

## Conclusiones


- Secuencia de reglas de decisin simples: una caracterstica y un umbral a la vez
- No se requiere escalar caractersticas numricas
- `max_depth` controla la compensacin entre el subajuste y el sobreajuste.
