## Redes neuronales con memoria

La red neuronal que hemos visto siempre devuelve la misma respuesta para las mismas entradas.
- No tiene memoria.
- En trminos de programacin, es un *algoritmo sin estado (stateless)*.



Para problemas ms complejos, como generar textos, necesitamos agregar estado a nuestro modelo.
- Cada vez que le pedimos a nuestra red neuronal una respuesta, tambin guardamos un conjunto de nuestros clculos intermedios y los reutilizamos la prxima vez como parte de nuestra entrada.
- De esa manera, nuestro modelo ajustar sus predicciones en funcin de la entrada que ha visto recientemente.



Esta es la idea bsica de una Red Neuronal Recurrente (Recurrent Neural Network).
- Estamos actualizando la red cada vez que la usamos.
- Esto le permite actualizar sus predicciones en funcin de lo que vio ms recientemente.
- Incluso puede modelar patrones con el tiempo siempre que le demos suficiente memoria.


Pero la parte realmente alucinante es que este algoritmo puede descubrir patrones en cualquier secuencia de datos.
- Puede generar fcilmente recetas de aspecto real o discursos falsos de Obama.
- Pero por qu limitarnos el lenguaje humano? --> Podemos aplicar esta misma idea a cualquier tipo de datos secuenciales que tengan un patrn.
