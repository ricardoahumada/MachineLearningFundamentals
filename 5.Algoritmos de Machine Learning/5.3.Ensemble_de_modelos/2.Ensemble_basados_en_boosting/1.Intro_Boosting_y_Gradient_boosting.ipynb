# Boosting y Gradient boosting

## Boosting para clasificacin



- Ajusta un primer rbol con depth = 1 y hace predicciones de error



- Esto impele a dar ms peso a los errores
- El segundo modelo intenta corregir los errores del anterior



- Vuelve a hacer lo mismo con el siguiente modelo



- Se sigue este proceso iterativamente hasta que se decide que se ha convergido

## Boosting para regresin

- Funciona de manera similar a la clasificacin, ajustando cada vez mejor, hasta converger.










## Boosting vs Gradient Boosting
- Boosting tradicional: `sklearn.ensemble.AdaBoostClassifier`
    - Muestras con mala prediccin son re-ponderadas en cada paso
    - Puede usar cualquier modelo base que acepte sample_weight

- Gradient Boosting: `sklearn.ensemble.HistGradientBoostingClassifier`
    - Cada modelo base predice el error negativo de los modelos anteriores
    - Sklearn usa rboles de decisin como modelo base.

### Gradient Boosting y caractersticas agrupadas

- `sklearn.ensemble.GradientBoostingClassifier`
    - Implementacin del mtodo tradicional (exacto)
    - Bueno para pequeos datasets
    - Muy lento para `n_samples > 10,000`

- `sklearn.ensemble.HistGradientBoostingClassifier`
    - Discretizar las caractersticas numricas (256 niveles)
    - Implementacin multi ncleo eficiente
    - **Mucho, mucho ms rpido** cuando n_samples es grande
