# Agrupacin (ensemble) de modelos basados en rboles

Los modelos basados en rboles son predictores poderosos individualmente, pero son un poco limitados.

- En particular, cuando los rboles son profundos, tienden a sobreajustar y si son demasiado poco profundos, tienden a subajustar.

Veremos cmo luchar contra esas debilidades construyendo agrupaciones para obtener modelos ms poderosos que se generalizan mejor.

Hay muchas estrategias para construir rboles de decisin.

- La primera estrategia se llama **bolsas (baggins) o bootstrap aggregating**, y puede refinarse an ms en los **bosques aleatorios (random forest)**.
- La segunda familia se llama **boosting** y puede refinarse an ms en los **rboles de decisin gradient boosted**.



Veremos que esas dos estrategias pueden usarse tanto para problemas de clasificacin como de regresin.

| Bagging | Boosting |
| --- | --- |
| fit trees independently | fit trees sequentially 
| each deep tree overfits  | each shallow tree underfits 
| averaging the tree predictions reduces overfitting  | sequentially adding trees reduces underfitting |


- Gradient boosting tiende a funcionar un poco mejor que bagging y random forest. 
- Los rboles poco profundos predicen ms rpido.
