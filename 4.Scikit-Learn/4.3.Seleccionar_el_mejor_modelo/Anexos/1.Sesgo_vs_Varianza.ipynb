# Sesgo (Bias) versus Varianza




**Sesgo alto == Subajuste:**
- Errores de prediccin sistemticos
- El modelo prefiere ignorar algunos aspectos de los datos
- Modelos mal especificados

**Alta varianza == Sobreajuste:**
- Errores de prediccin sin una estructura obvia
- Pequeo cambio en el conjunto de entrenamiento, gran cambio en el modelo
- Modelos inestables

> El sesgo puede provenir de la eleccin de la familia modelo.
